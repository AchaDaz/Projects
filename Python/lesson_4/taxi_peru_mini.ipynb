```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
taxi= pd.read_csv(r'C:\Users\user\Documents\Karpov Courses\Python\lesson_3\3_taxi_peru.csv', sep=';')
```

```python
times = ['start_at', 'end_at', 'arrived_at']
for time in times:
    taxi[time] = pd.to_datetime(taxi[time])
```

```python
types = ['asap', 'reserved']
taxi = taxi.query('start_type in @types')
```

```python
#Создаем колонку со значением разницы между временем заказа и временем подачи машины в минутах
taxi.head()
```

```python
taxi['wait_time'] = (taxi['arrived_at']-taxi['start_at']).astype('timedelta64[m]')
```

```python
taxi.head()
```

```python
taxi.query('start_type == "reserved" and wait_time>0.0') \
    .groupby('driver_id', as_index=True) \
    .agg({'driver_score' : 'count'}) \
    .idxmax()
```

```python
taxi.head()
```

```python
taxi['month'] = taxi['start_at'].dt.month
```

```python
taxi['weekday'] = taxi['start_at'].dt.day_name()
```

```python
#График зависимости количества заказов от месяца
plt.figure(figsize=(12,8))
ax = sns.countplot(x='month', data=taxi)
```

```python
taxi.weekday.unique()
```

```python
#График зависимости количества заказов от дня недели
plt.figure(figsize=(12,8))
ax = sns.countplot(x='weekday', data=taxi, order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
```

```python
#MAU (monthly active users) – число уникальных пользователей за месяц
plt.plot(taxi.groupby('month', as_index=False).nunique('user_id')['month'], taxi.groupby('month', as_index=False).nunique('user_id')['user_id'])
```

```python
file = open('gff.tsv')
```

```python
DNA = file.readlines()
```

```python
repr(DNA[2])
```

```python
for i in range(len(DNA)):
    DNA[i] = DNA[i].split(';')
```

```python
DNA
```

```python
for i in range(len(DNA)):
    DNA[i] = re.split(';|=|\n|,|\t',DNA[i])
```

```python
DNA
```

```python
seq_id = []
source = []
types = []
for line in DNA:
    seq_id.append(line[0])
    source.append(line[1])
    types.append(line[2])
    
seq_id.remove('seq_id')
source.remove('source')
types.remove('type')
```

```python
list_dna = []
for line in DNA:
    for i in range(len(line)):
        if line[i].startswith('XM'):
            list_dna.append(line[i])        
```

```python
dna_df = pd.DataFrame(list_dna, columns=['Parent'])
```

```python
dna_df
```

```python
dna_df.value_counts()
```

```python
dna_df['seq_id'] = seq_id
dna_df['source'] = source
dna_df['types'] = types
```

```python
dna_df.types.value_counts()
```

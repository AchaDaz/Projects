```python
taxi_df = taxi_df.rename(columns={'pcp 01': 'pcp_01',
                                  'pcp 06': 'pcp_06',
                                  'pcp 24': 'pcp_24'})
```

```python
taxi_df.head()
```

```python
taxi_df.borough.value_counts()
```

```python
taxi_df.pickups.sum()
```

```python
taxi_df.groupby('borough') \
    .agg({'pickups' : 'sum'}) \
    .sort_values('pickups', ascending=False)
```

```python
taxi_df.pickups.idxmax()
```

```python
a = taxi_df.groupby('borough') \
    .agg({'pickups' : 'sum'}) \
    .idxmin()
print(a)
```

```python
taxi_df.groupby(['borough', 'hday']).agg({'pickups': 'sum', 'pickups': 'mean'})
```

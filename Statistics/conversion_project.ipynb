```python
conversion_fb = conversion_df.groupby('fb_campaign_id', as_index=False) \
                             .agg({'Impressions':'sum'})
```

### CTR (click-through rate)

```python
conversion_df['ctr'] = conversion_df.Clicks/conversion_df.Impressions
```
